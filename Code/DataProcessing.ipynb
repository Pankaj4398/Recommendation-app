# Merge

```python
datasheet = pd.merge(data_recommend, product_recommend, on = 'itemid')
datasheet.shape
```

```python
datasheet.head(2)
```

```python
datasheet = datasheet.drop(columns=['orderItems', 'description'])
```

```python
datasheet.head()
datasheet.to_csv('reviews.csv')
```

```python
dataset = datasheet.pivot_table(index='itemid',columns='userid',values='rating')
```

```python
dataset.head()
```

```python
dataset.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
csr_dataset = csr_matrix(dataset.values)
dataset.reset_index(inplace=True)
```

```python
dataset.head()
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=20, n_jobs=-1)
model.fit(csr_dataset)
```

```python
import pickle
filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model
print(result)
```

```python
def food_recommendation(Food_Name):
    n = 10
    FoodList = datasheet[datasheet['name'].str.contains(Food_Name)]  
    if len(FoodList):        
        Foodi= FoodList.iloc[0]['itemid']
        Foodi = dataset[dataset['itemid'] == Foodi].index[0]
        distances , indices = model.kneighbors(csr_dataset[Foodi],n_neighbors=n+1)    
        Food_indices = sorted(list(zip(indices.squeeze().tolist(),distances.squeeze().tolist())),key=lambda x: x[1])[:0:-1]
        Recommendations = []
        for val in Food_indices:
            Foodi = dataset.iloc[val[0]]['itemid']
            i = datasheet[datasheet['itemid'] == Foodi].index
            Recommendations.append({'name':datasheet.iloc[i]['name'].values[0],'Distance':val[1]})
        df = pd.DataFrame(Recommendations,index=range(1,n+1))
        return df['name']
    else:
        return "No Similar Foods."
```
